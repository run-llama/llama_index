```python
from imap_tools import MailBox
from typing import (Any, Dict)
from llama_index.core.schema import Document
from dotenv import load_dotenv
load_dotenv()
import os

server = MailBox(os.getenv("MAIL_HOST"))
server.login(os.getenv("MAIL_USERNAME"), os.getenv("MAIL_PASSWORD"))

server.folder.set("INBOX")

metadata_keys = ["uid", "subject", "from_values", "to_values"]

for msg in server.fetch():
    doc_metadata: Dict[str, Any] = {}
    for key in metadata_keys:
        doc_metadata[key] = getattr(msg, key, None)

    text = f"From: {msg.from_}, To: {msg.to[0]}, Subject: {msg.subject}, Message: {msg.text}"
    print(Document(text=text, metadata=doc_metadata))
```
