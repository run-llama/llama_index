```python
from llama_index.memory.bedrock_agentcore import AgentCoreMemory, AgentCoreMemoryContext

context = AgentCoreMemoryContext(
    actor_id="test_users_1",
    memory_id="memory_rwtqe-ju7bYH7CBj",
    session_id="test-store-3-session-7",
    memory_strategy_id="memory_semantic_rwtqe-fJGJOoGFkh"
)
agentcore_memory = AgentCoreMemory(
    context=context
)

# print(agentcore_memory.get())
```

```python
from llama_index.llms.bedrock_converse import BedrockConverse
from llama_index.core.agent.workflow import FunctionAgent
from llama_index.tools.yahoo_finance import YahooFinanceToolSpec

llm = BedrockConverse(model="us.anthropic.claude-sonnet-4-20250514-v1:0")

def call_fn(name: str):
    """
    Call the provided name.

    Args:
        name: str (Name of the person)

    """
    print(f"Calling... {name}")


def email_fn(name: str):
    """
    Email the provided name.

    Args:
        name: str (Name of the person)

    """
    print(f"Emailing... {name}")

finance_tool_spec = YahooFinanceToolSpec()
agent = FunctionAgent(
    tools=finance_tool_spec.to_tool_list(),
    llm=llm,
)
```

```python
# Try to break it by asking for messages that are older than 12 msgs
# Ask for a previous stock that I asked before to prove that it's pulling from memory.
 # Create a new session and ask about a question from an older session to ensure that it's isolated

import time

start_time = time.time()
response = await agent.run("What are my favorite musicians?", memory=agentcore_memory)
end_time = time.time()
execution_time = end_time - start_time

print("\n" + str(response))
print(f"\nExecution time: {execution_time:.2f} seconds")
```
