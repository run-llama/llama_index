## 1. Basic Web Search

````python
from llama_index.tools.searchapi import SearchApiToolSpec

tool_spec = SearchApiToolSpec()

docs = tool_spec.search("LlamaIndex RAG 2025", num=5)
for doc in docs:
    print(doc.extra_info["position"], doc.extra_info["title"])
    print("  ", doc.extra_info["link"])
    print("  ", doc.text[:120])
    print()
````

## 2. News Search

````python
news_docs = tool_spec.news_search("AI regulation EU 2025", num=5)
for doc in news_docs:
    print(doc.extra_info["title"])
    print("  Source:", doc.extra_info["source"], "|", doc.extra_info["date"])
    print("  ", doc.text[:120])
    print()
````

## 3. Academic / Scholar Search

````python
scholar_docs = tool_spec.scholar_search("retrieval augmented generation survey", num=5)
for doc in scholar_docs:
    print(doc.extra_info["title"])
    print("  Publication:", doc.extra_info["publication_info"])
    print("  Cited by:", doc.extra_info["cited_by_count"])
    print("  ", doc.text[:120])
    print()
````

## 4. Image Search

````python
image_docs = tool_spec.image_search("neural network diagram", num=3)
for doc in image_docs:
    print(doc.text)  # image title
    print("  Full-size:", doc.extra_info["original"])
    print()
````

## 5. Use With an OpenAI Agent

Convert the spec to a tool list and wire it into an agent.

````python
from llama_index.agent.openai import OpenAIAgent
from llama_index.llms.openai import OpenAI

tools = tool_spec.to_tool_list()

agent = OpenAIAgent.from_tools(
    tools,
    llm=OpenAI(model="gpt-4o"),
    verbose=True,
)

response = agent.chat(
    "Search for the top 3 most-cited academic papers on "
    "retrieval-augmented generation and summarise their findings."
)
print(response)
````
