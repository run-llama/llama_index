# DaoXE (OpenAI-compatible gateway)

This notebook shows [OpenAILike](https://docs.llamaindex.ai/) pointed at [DaoXE](https://daoxe.com),
a multi-model multi-protocol API gateway.

Set `DAOXE_API_KEY` and `DAOXE_MODEL` (exact account model ID from `GET /v1/models`).
Disclosure: example maintained by a DaoXE maintainer. Not available in mainland China.


```python
%pip install llama-index-llms-openai-like
```

```python
import os
from llama_index.llms.openai_like import OpenAILike

llm = OpenAILike(
    model=os.environ["DAOXE_MODEL"],
    api_base="https://daoxe.com/v1",
    api_key=os.environ["DAOXE_API_KEY"],
    is_chat_model=True,
    is_function_calling_model=True,
    context_window=128000,
)

resp = llm.complete("Reply with OK")
print(resp)
```
