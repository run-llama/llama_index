# FuturMix

FuturMix is a unified AI gateway that provides access to 22+ models from OpenAI, Anthropic, Google, and more through a single OpenAI-compatible API endpoint. You can find out more on their [homepage](https://futurmix.ai).

Since FuturMix exposes an OpenAI-compatible API, you can use LlamaIndex's `OpenAILike` class to interact with it directly.

Visit [futurmix.ai](https://futurmix.ai) to sign up and get an API key.

```python
from llama_index.llms.openai_like import OpenAILike
from llama_index.core.llms import ChatMessage
```

## Call `chat` with ChatMessage List
You need to set your FuturMix API key. You can either set the env var `FUTURMIX_API_KEY` or pass it directly in the constructor.

```python
import os

# os.environ["FUTURMIX_API_KEY"] = "<your-futurmix-key>"

llm = OpenAILike(
    model="gpt-4o",
    api_base="https://futurmix.ai/v1",
    api_key="<your-futurmix-key>",
    is_chat_model=True,
)
```

```python
message = ChatMessage(role="user", content="Tell me a joke")
resp = llm.chat([message])
print(resp)
```

### Streaming

```python
message = ChatMessage(role="user", content="Tell me a story in 250 words")
resp = llm.stream_chat([message])
for r in resp:
    print(r.delta, end="")
```

## Call `complete` with Prompt

```python
resp = llm.complete("Tell me a joke")
print(resp)
```

```python
resp = llm.stream_complete("Tell me a story in 250 words")
for r in resp:
    print(r.delta, end="")
```

## Model Configuration

FuturMix supports models from multiple providers through a single API key:

| Provider | Example Models |
|----------|---------------|
| OpenAI | `gpt-4o`, `gpt-4o-mini` |
| Anthropic | `claude-sonnet-4-20250514`, `claude-3-5-haiku-20241022` |
| Google | `gemini-2.5-flash-preview-04-17`, `gemini-2.0-flash` |

View the full list of available models at [futurmix.ai](https://futurmix.ai).

```python
# Using Claude
llm = OpenAILike(
    model="claude-sonnet-4-20250514",
    api_base="https://futurmix.ai/v1",
    api_key="<your-futurmix-key>",
    is_chat_model=True,
)
```

```python
resp = llm.complete("Explain quantum computing in simple terms")
print(resp)
```

```python
# Using Gemini
llm = OpenAILike(
    model="gemini-2.5-flash-preview-04-17",
    api_base="https://futurmix.ai/v1",
    api_key="<your-futurmix-key>",
    is_chat_model=True,
)
```

```python
resp = llm.complete("Write a haiku about programming")
print(resp)
```
