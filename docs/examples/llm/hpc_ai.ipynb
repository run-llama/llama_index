# HPC-AI

[HPC-AI](https://api.hpc-ai.com/) provides an OpenAI-compatible inference API. Example model IDs:

- `minimax/minimax-m2.5`
- `moonshotai/kimi-k2.5`

Default base URL: `https://api.hpc-ai.com/inference/v1`.

````python
from llama_index.llms.hpc_ai import HpcAiLLM
````

````python
llm = HpcAiLLM(
    model="minimax/minimax-m2.5",
    api_key="your_api_key",
)
````

## Completion

````python
response = llm.complete("Say hello in one short sentence.")
````

````python
print(response)
````

## Chat

````python
from llama_index.core.llms import ChatMessage

messages = [
    ChatMessage(role="user", content="What is 2+2? Reply with just the number."),
]
resp = llm.chat(messages)
````

````python
print(resp)
````
