# Foundry Local

[Foundry Local](https://github.com/microsoft/Foundry-Local) from Microsoft exposes an OpenAI-compatible API for running local models. LlamaIndex can use it directly through the `OpenAILike` integration.

## Setup

1. Install `llama-index-llms-openai-like` and the Foundry Local SDK (`foundry-local-sdk-winml` on Windows, `foundry-local-sdk` on macOS/Linux).
2. Use the Foundry Local SDK to download/load a model and start the optional local OpenAI-compatible web service.
3. `OpenAILike` needs an HTTP endpoint, so point it at `f"{manager.urls[0]}/v1"` and use `model.id` so requests target the exact loaded variant.
4. Foundry Local itself is SDK-first and does not require the web service outside OpenAI-compatible integrations.
5. Increase `timeout` if your hardware takes longer to generate the first response.

```python
# On Windows, replace foundry-local-sdk with foundry-local-sdk-winml
%pip install llama-index-llms-openai-like foundry-local-sdk
```

```python
from foundry_local_sdk import Configuration, FoundryLocalManager
from llama_index.llms.openai_like import OpenAILike
```

```python
FoundryLocalManager.initialize(Configuration(app_name="llamaindex-foundry-local"))
manager = FoundryLocalManager.instance

model = manager.catalog.get_model("qwen2.5-0.5b")
model.download()
model.load()
manager.start_web_service()

llm = OpenAILike(
    model=model.id,
    api_base=f"{manager.urls[0].rstrip('/')}/v1",
    api_key="fake",
    context_window=32768,
    is_chat_model=True,
    is_function_calling_model=False,
    timeout=120.0,
)
```

```python
response = llm.complete("Reply with: Foundry Local is working.")
print(str(response))
```
